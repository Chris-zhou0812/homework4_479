4.Initial Column Clean up

* Column LOCATION shall be deleted because there are column LONGITUDE and LATITUDE that contains the same(even more exact) data.

```python
# count that LONGITUDE contains the same with that in LOCATION is the same as the size of dataset
count = 0
for i in range(len(dfn['LONGITUDE'])):
    if str(dfn['LONGITUDE'][i]) == "0.0":
        count += 1
    else:
        if str(dfn['LONGITUDE'][i])[:3] in str(dfn['LOCATION'][i]):
            count += 1
print(count)
```

```python
# count that LATITUDE contains the same with that in LOCATION is the same as the size of dataset
count = 0
for i in range(len(dfn['LATITUDE'])):
    if str(dfn['LATITUDE'][i]) == "0.0":
        count += 1
    else:
        if str(dfn['LATITUDE'][i])[:3] in str(dfn['LOCATION'][i]):
            count += 1
print(count)
```
